```python
file_path = 'airports.dat'
airport_names = []

try:
    with open(file_path, 'r') as f:
        for line in f:
            airport_data = line.strip().split(',')
            airport_name = airport_data[1].strip('"')
            airport_names.append(airport_name)

    num_airport_names = len(airport_names)
    print("Number of airport names:", num_airport_names)

except FileNotFoundError:
    print(f"File '{file_path}' not found.")
```

```python
file_path = 'airports.dat'
australian_airports = []

try:
    with open(file_path, 'r') as f:
        for line in f:
            airport_data = line.strip().split(',')
            country = airport_data[3].strip('"')
            if country == 'Australia':
                australian_airports.append(airport_data)

    num_australian_airports = len(australian_airports)
    print("Number of Australian airports:", num_australian_airports)

except FileNotFoundError:
    print(f"File '{file_path}' not found.")
```

```python
file_path = 'airports.dat'
russian_airports = []

try:
    with open(file_path, 'r') as f:
        for line in f:
            airport_data = line.strip().split(',')
            country = airport_data[3].strip('"')
            if country == 'Russia':
                russian_airports.append(airport_data)

    num_russian_airports = len(russian_airports)
    print("Number of Russian airports:", num_russian_airports)

except FileNotFoundError:
    print(f"File '{file_path}' not found.")
```
